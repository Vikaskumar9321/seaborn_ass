Seaborn is a popular data visualization library in Python that provides a high-level interface for creating information and aesthetically pleasing plots. Here are five common plots that can be created using Seaborn,
along with their typical uses:

1.) Histogram: A histogram is used to represent the distribution of a single numerical variable. it divides the variable into bins and displays the count of density of observations in each bin. Histograms are useful for understanding the underlying data distribution and identifying any outliers or skewness.

2.) Scatter plot: A scatter plot displays the relatioship between wo numerical variables. It represents each data point as a marker on a two-dimensional plane, with one variable on the x-axis and the other on the y-axis. scatter plots are useful for visualization patterns correlations, and outliers in data.

3.) Box plot: A box plot, also known as a whisker plot, is used to dispaly the summary statistics of a numerical variable. it shows the median quartilers, and any outliers in the data. Bos plots are helpful for comparing distributions, identifying skewness, and detecting potential outliers.

4.) Bar Plot: A bar plot is commonly used to represent categorical data. it displays the frequency or proportion of each categories or groups and visulaizing the distribution of categorical variables.

5.) Heatmap: A heatmap is a graphical represnetation of data where values are displayed as colors in a grid, It is often used to visualize the correlation or relationship between multiple variables.Heatmaps are useful for identifying patterns, clusters and relationships in large datasets

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('fmri.csv')
sns.lineplot(data=df, x='timepoint', y='signal', hue='event', style='region')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.title('FMRI Signal by Events and Regions')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_dataset = sns.load_dataset('titanic')
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
sns.boxplot(data = titanic_dataset, x = 'pclass', y = 'age')
plt.title('Box plot of age by passenger class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

plt.subplot(1,2,2)
sns.boxplot(data=titanic_dataset, x='pclass', y='fare')
plt.title('Box plot of Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.tight_layout()
plt.show()
```

```python
diamonds_dataset = sns.load_dataset('diamonds')
sns.set(style='darkgrid' )
plt.figure(figsize=(10,6))
sns.histplot(data=diamonds_dataset, x='price', hue='cut', multiple ='stack')
plt.title("Histogram of price with cut")
plt.xlabel('Price')
plt.ylabel('Count')
plt.legend(title='Cut')
plt.show()
```

```python
flights = sns.load_dataset("flights")
flights_matrix = flights.pivot("month","year","passengers")
sns.heatmap(flights_matrix,annot = True, fmt="d", cmap ="YlGnBu")
plt.title("Passenger number (1000s)")
plt.show()
```
